```python
np.array:
Purpose: Creates a new NumPy array.
Behavior: Always creates a copy of the input data, unless copy=False is explicitly specified.
Use Case: When you need a new array and want to ensure that the data is copied.

e.g
import numpy as np
list_data = [1, 2, 3]
array = np.array(list_data)

np.asarray:
Purpose: Converts the input to a NumPy array.
Behavior: Does not create a copy if the input is already an ndarray with the same dtype.
Use Case: When you want to avoid unnecessary copying and work with the original data if it’s already an array.

e.g
import numpy as np
list_data = [1, 2, 3]
array = np.asarray(list_data)

np.asanyarray:
Purpose: Converts the input to a NumPy array, but passes through ndarray subclasses.
Behavior: Similar to np.asarray, but if the input is a subclass of ndarray (like a masked array), it returns the subclass.
Use Case: When you want to preserve the type of ndarray subclasses.

e.g
import numpy as np
list_data = [1, 2, 3]
array = np.asanyarray(list_data)


 Difference between Deep Copy and Shallow Copy
Shallow Copy:
Definition: Creates a new object, but inserts references into it to the objects found in the original.
Behavior: Changes to the original object may affect the shallow copy if the changes are made to mutable objects within the original.
Use Case: When you need a copy of an object but don’t need to duplicate the nested objects.

e.g
import copy
original_list = [[1, 2, 3], [4, 5, 6]]
shallow_copy = copy.copy(original_list)

Deep Copy:
Definition: Creates a new object and recursively copies all objects found in the original.
Behavior: Changes to the original object do not affect the deep copy.
Use Case: When you need a completely independent copy of an object, including all nested objects.

e.g
import copy
original_list = [[1, 2, 3], [4, 5, 6]]
deep_copy = copy.deepcopy(original_list)
```

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a time-series dataset
date_range = pd.date_range(start='2024-01-01', periods=100, freq='D')
temperature = np.random.uniform(15, 35, size=100)
humidity = np.random.uniform(40, 80, size=100)

data = {
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
}

df = pd.DataFrame(data)

# Plotting Temperature and Humidity on the same plot with different y-axes
fig, ax1 = plt.subplots()

ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='tab:red')
ax1.plot(df['Date'], df['Temperature'], color='tab:red')
ax1.tick_params(axis='y', labelcolor='tab:red')

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
ax2.set_ylabel('Humidity (%)', color='tab:blue')
ax2.plot(df['Date'], df['Humidity'], color='tab:blue')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Create a NumPy array with 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Step 2: Plot a histogram of the data with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='b', label='Histogram')

# Step 3: Overlay a line plot representing the normal distribution's PDF
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, data.mean(), data.std())
plt.plot(x, p, 'k', linewidth=2, label='PDF')

# Step 4: Label the x-axis and y-axis
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# Step 5: Set the title of the plot
plt.title('Histogram with PDF Overlay')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Generate two random arrays
x = np.random.randn(100)
y = np.random.randn(100)

# Determine the quadrant for each point
def get_quadrant(x, y):
    if x >= 0 and y >= 0:
        return 'Q1'
    elif x < 0 and y >= 0:
        return 'Q2'
    elif x < 0 and y < 0:
        return 'Q3'
    else:
        return 'Q4'

# Create a DataFrame
data = pd.DataFrame({'x': x, 'y': y})
data['Quadrant'] = data.apply(lambda row: get_quadrant(row['x'], row['y']), axis=1)

# Create the scatter plot
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(data=data, x='x', y='y', hue='Quadrant', palette='viridis', s=100)

# Add legend, labels, and title
plt.legend(title='Quadrant')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Show the plot
plt.show()
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Generate data for the sine wave
x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

# Output to notebook
output_notebook()

# Create a new plot
p = figure(title="Sine Wave Function", x_axis_label='X-axis', y_axis_label='Y-axis')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="Sine Wave", line_width=2)

# Add grid lines
p.xgrid.grid_line_color = 'gray'
p.ygrid.grid_line_color = 'gray'

# Show the results
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np
import pandas as pd

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 100, size=len(categories))

# Create a DataFrame
data = pd.DataFrame({'categories': categories, 'values': values})

# Create a ColumnDataSource
source = ColumnDataSource(data)

# Output to notebook
output_notebook()

# Create a new plot
p = figure(x_range=categories, title="Random Categorical Bar Chart", x_axis_label='Categories', y_axis_label='Values')

# Add bars
p.vbar(x='categories', top='values', width=0.9, source=source, legend_field="categories",
       line_color='white', fill_color='navy')

# Add hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Show the results
show(p)
```
